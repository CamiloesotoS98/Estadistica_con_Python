# Histograma

```python
tiempos_respuesta = [1.2, 2.8, 2.1, 3.6, 3.9, 4.0, 5.1, 5.5, 6.7, 7.3, 7.6, 8.2, 8.4, 9.1, 9.9]
fig, ax = plt.subplots()
n, bins, patches = ax.hist(tiempos_respuesta, bins=6,color='blue', edgecolor='black') # Cantidad de divisiones
ax.bar_label(patches)
ax.set_title('Histograma de Tiempos de Respuesta')
ax.set_xlabel('Tiempo (segundos)')
ax.set_ylabel('Frecuencia')
plt.show()
# ancho de la barra=  (máximo - mínimo)/ bins
# ancho de la barra=(9.9-1.2)/6 = 1.45
```

```python
def crear_tabla_frecuencias(valores):
    frecuencia_absoluta = pd.Series(valores).value_counts().sort_index()
    frecuencia_absoluta_acumulada = frecuencia_absoluta.cumsum()
    frecuencia_relativa = frecuencia_absoluta / len(valores)
    frecuencia_relativa_acumulada = frecuencia_relativa.cumsum()
    tabla_frecuencias = pd.DataFrame({
        'Frecuencia Absoluta': frecuencia_absoluta,
        'Frecuencia Absoluta Acumulada': frecuencia_absoluta_acumulada,
        'Frecuencia Relativa': frecuencia_relativa,
        'Frecuencia Relativa Acumulada': frecuencia_relativa_acumulada
    })
    return tabla_frecuencias

valores = ['Coca-Cola', 'Pepsi', 'Sprite', 'Cuatro', 'Colombiana', 'Coca-Cola', 'Pepsi', 'Pepsi', 'Sprite', 'Cuatro']
tabla = crear_tabla_frecuencias(valores)
tabla
```

```python
import matplotlib.pyplot as plt
import pandas as pd


Mes = ['Enero', 'Enero', 'Enero', 'Febrero', 'Febrero', 'Febrero', 'Marzo', 'Marzo', 'Marzo']
Refresco = ['Coca-Cola', 'Pepsi', 'Sprite', 'Coca-Cola', 'Pepsi', 'Sprite', 'Coca-Cola', 'Pepsi', 'Sprite']
Cantidad = [30, 20, 10, 25, 15, 5, 35, 25, 15]

data = {'Mes': Mes, 'Refresco': Refresco, 'Cantidad': Cantidad}
df = pd.DataFrame(data)

df_pivot = df.pivot(index='Mes', columns='Refresco', values='Cantidad')
ax = df_pivot.plot(kind='bar', stacked=True)


ax.set_title('Ventas de Refrescos por Mes')
ax.set_xlabel('Mes')
ax.set_ylabel('Cantidad Vendida')
ax.legend(title='Refresco')
plt.show()

```

```python
data = {
    'Mes': ['Enero', 'Enero', 'Enero', 'Febrero', 'Febrero', 'Febrero', 'Marzo', 'Marzo', 'Marzo'],
    'Refresco': ['Coca-Cola', 'Pepsi', 'Sprite', 'Coca-Cola', 'Pepsi', 'Sprite', 'Coca-Cola', 'Pepsi', 'Sprite'],
    'Cantidad': [30, 20, 10, 25, 15, 5, 35, 25, 15]
}

df = pd.DataFrame(data)
contingency_table = pd.crosstab(index=df['Mes'], columns=df['Refresco'], values=df['Cantidad'], aggfunc='sum').fillna(0)
mosaic(contingency_table.stack(), title='Gráfico de Mosaicos de Ventas de Refrescos por Mes')
plt.xlabel('Mes')
plt.ylabel('Cantidad Vendida')
plt.show()
```

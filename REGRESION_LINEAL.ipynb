```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr
```

```python
X = [-3,-1,0,2]
Y=[0,1,2,3]
plt.plot(X, Y, marker='o', linestyle='-')
plt.title('Diagrama de dispersión')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
m, c = np.polyfit(X, Y, 1)
# Mostrar los resultados
print(f"Pendiente (m): {m}")
print(f"Intersección (c): {c}")
```

```python
# Generar la recta de regresión
X_regresion = np.linspace(min(X), max(X), 100)
Y_regresion = m * X_regresion + c

# Graficar los datos y la recta de regresión
plt.scatter(X, Y, color='blue', label='Datos')
plt.plot(X_regresion, Y_regresion, color='red', label='Recta de Regresión')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
X = [4,6,11,3,16]
Y=[50,50,40,60,30]
```

```python
correlation_coefficient, p_value = pearsonr(X, Y)

# Mostrar los resultados
print(f"Coeficiente de correlación de Pearson: {correlation_coefficient}")
print(f"Valor p: {p_value}")
```

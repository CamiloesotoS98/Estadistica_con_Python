```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from scipy.stats import t
```

# INTERVALO DE CONFIANZA CON DESVIACIÓN ESTÁNDAR CONOCIDA

```python
desviacion_estandar_poblacional = 6  # Varianza de la población (σ^2)
nivel_confianza = 0.99  # Nivel de confianza
media_muestral = 32  # Media muestral (x̄)
n = 50  # Tamaño de la muestra (n)

# Calculamos el error estándar de la media (σ_x̄)
error_estandar_media = desviacion_estandar_poblacional / np.sqrt(n)

# Calculamos el valor crítico z para el nivel de confianza dado
z_critico = stats.norm.ppf((1 + nivel_confianza) / 2)

# Calculamos el margen de error
margen_error = z_critico * error_estandar_media

# Calculamos el intervalo de confianza
limite_inferior = media_muestral - margen_error
limite_superior = media_muestral + margen_error

print("Intervalo de confianza del 95%:")
print("Límite inferior:", limite_inferior)
print("Límite superior:", limite_superior)
```

```python
x = np.linspace(-5, 5, 400)

df1 = 1
df2 = 95
y1 = t.pdf(x, df1)
y2 = t.pdf(x, df2)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label=f't-distribution (df={df1})', color='blue')
plt.plot(x, y2, label=f't-distribution (df={df2})', color='red')
plt.title('Distribución t con diferentes grados de libertad')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)

plt.show()
```

# INTERVALO DE CONFIANZA CON DESVIACIÓN ESTÁNDAR DESCONOCIDA

```python
desviacion_estandar_muestral = 4007  # Desviación estándar muestral (s)
nivel_confianza = 0.95  # Nivel de confianza
media_muestral = 9312  # Media muestral (x̄)
n = 70  # Tamaño de la muestra (n)

# Calculamos el error estándar de la media (s_x̄)
error_estandar_media = desviacion_estandar_muestral / np.sqrt(n)

# Calculamos el valor crítico t para el nivel de confianza dado y n-1 grados de libertad
t_critico = stats.t.ppf((1 + nivel_confianza) / 2, df=n-1)

# Calculamos el margen de error
margen_error = t_critico * error_estandar_media

# Calculamos el intervalo de confianza
limite_inferior = media_muestral - margen_error
limite_superior = media_muestral + margen_error

print(f"Intervalo de confianza del {nivel_confianza*100}%:")
print("Límite inferior:", limite_inferior)
print("Límite superior:", limite_superior)
```

# TAMAÑO DE MUESTRA

```python
estimacion_desviacion_estandar = 9.65  # Estimación de la desviación estándar (s)
nivel_confianza = 0.95  # Nivel de confianza
margen_error = 2  # Margen de error deseado (E)

# Calculamos el valor crítico z para el nivel de confianza dado
z_critico = stats.norm.ppf((1 + nivel_confianza) / 2)

# Calculamos el tamaño de la muestra
n = ( (z_critico * estimacion_desviacion_estandar) / margen_error )**2
#n = np.ceil(n)  # Redondeamos al siguiente entero más alto

print(f"Tamaño de muestra necesario para un nivel de confianza del {nivel_confianza*100}% y un margen de error de {margen_error}: {n}")
```
